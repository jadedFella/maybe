```python
import pandas as pd
import numpy as np
import math
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

#Importing dataset
X=np.array([1,2,3,4,5])
Y=np.array([3,4,5,7,5])

#Gradient Descent function
def gradient_descent(X, Y, learning_rate, iterations):
    m = 0
    c = 0
    n = len(X)
    for _ in range(iterations):
        Y_pred = m*X + c
        D_m = (1/n) * sum(X * (Y_pred - Y))
        D_c = (1/n) * sum(Y_pred - Y)
        m = m - learning_rate * D_m
        c = c - learning_rate * D_c
    return m, c

#Applying Gradient Descent
learning_rate=0.01
iterations=1000
m, c = gradient_descent(X, Y, learning_rate, iterations)
#Predicting the values
Y_pred = m*X + c

#Printing the accuracy results
print("RSME = " , math.sqrt(mean_squared_error(Y,Y_pred)))
print("MAE = ", mean_absolute_error(Y,Y_pred))
print("R2_score = " , r2_score(Y,Y_pred))

#Plotting the graph
plt.scatter(X, Y, color='blue')
plt.plot(X, Y_pred, color='red')
plt.title('Gradient Descent')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

#Implementing Ridge regression
from sklearn.linear_model import Ridge
regressor = Ridge(alpha=0.01, max_iter=1000)
X=X.reshape(-1,1)
regressor.fit(X, Y)

#Predicting the values
Y_pred = regressor.predict(X)

#Printing the results
print("RSME = " , math.sqrt(mean_squared_error(Y,Y_pred)))
print("MAE = ", mean_absolute_error(Y,Y_pred))
print("R2_score = " , r2_score(Y,Y_pred))

#Plotting the graph
plt.scatter(X, Y, color='blue')
plt.plot(X, Y_pred, color='red')
plt.title('Ridge Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
